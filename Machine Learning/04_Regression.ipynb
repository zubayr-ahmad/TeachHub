### üß† What is Regression?

Regression is a type of **Supervised Learning** technique used to predict a continuous output variable (target) based on one or more input features.

- **Example Use Cases**:
  - Predicting house prices based on area, location, etc.
  - Estimating salary based on years of experience.
  - Forecasting sales, temperature, or stock prices.

---

### üìä Regression vs Classification

| Aspect | Regression | Classification |
|--------|------------|----------------|
| Output Type | Continuous | Categorical |
| Examples | Salary, Age, House Price | Spam/Not Spam, Disease/No Disease |

---

### üìê Linear Regression (Simple)

Let‚Äôs start with the most basic form of regression:

**Linear Regression Equation:**

\[
y = m \cdot x + c
\]

Where:
- `y` is the predicted value.
- `x` is the input feature.
- `m` is the slope (weight/parameter).
- `c` is the y-intercept (bias/constant).

---

### üîç How ML Learns `m` and `c`

We try different values of `m` and `c` to minimize the **error** between predicted and actual values.

We measure this error using a **Cost Function**.

---

### üìâ Cost Function (Mean Squared Error)

\[
   J(m, c) = \frac{1}{n} \sum_{i=1}^{n}(y_{pred}^{(i)} - y_{true}^{(i)})^2
\]

This function tells us how well our model is performing. The smaller the cost, the better the model.

---

### üí° Problem: How do we find the best `m` and `c`?

We want to find the values of `m` and `c` that minimize the cost function. One way is to use a technique called **Gradient Descent**.

---

## üîΩ Gradient Descent in Regression

### üìò What is Gradient Descent?

Gradient Descent is an **optimization algorithm** used to find the minimum of a function.

In Regression, we use it to minimize the **Cost Function (MSE)** by updating parameters `m` and `c` step by step.

---

### ‚öôÔ∏è How does it work?

1. Initialize random values for `m` and `c`.
2. Calculate predictions using current `m` and `c`.
3. Compute the cost (MSE).
4. Calculate the gradients:
   \[
   \frac{\partial J}{\partial m} = \frac{2}{n} \sum (y_{pred} - y_{true}) \cdot x
   \]
   \[
   \frac{\partial J}{\partial c} = \frac{2}{n} \sum (y_{pred} - y_{true})
   \]
5. Update parameters:
   \[
   m = m - \alpha \cdot \frac{\partial J}{\partial m}
   \]
   \[
   c = c - \alpha \cdot \frac{\partial J}{\partial c}
   \]
   Where `Œ±` is the **Learning Rate** ‚Äì controls step size.
6. Repeat steps 2-5 until the cost function converges (reaches minimum).

---

### ‚úçÔ∏è Step-by-Step Example

Let‚Äôs say we have this dataset:

| Experience (Years) | Salary (in $1000s) |
|--------------------|--------------------|
| 1 | 30 |
| 2 | 35 |
| 3 | 40 |
| 4 | 45 |

#### Step 1: Initialize
Let `m = 0`, `c = 0`, and `learning_rate = 0.01`

#### Step 2: Predict
\[
y_{pred} = m \cdot x + c
\]

#### Step 3: Compute Cost (MSE)

#### Step 4: Compute Gradients

\[
\frac{\partial J}{\partial m} = \frac{2}{n} \sum (y_{pred} - y_{true}) \cdot x
\]

\[
\frac{\partial J}{\partial c} = \frac{2}{n} \sum (y_{pred} - y_{true})
\]

#### Step 5: Update m and c

Repeat until convergence.

---

### üìà Visualizing Gradient Descent

- Imagine the cost function as a **bowl-shaped curve**.
- Gradient Descent moves **downhill** step-by-step to reach the lowest point of the curve (global minimum).

---

### üìå Importance of Learning Rate (Œ±)
- Too high ‚Üí model jumps and never converges.
- Too low ‚Üí model takes forever to reach the minimum.

---

### üìé Python Implementation of Gradient Descent (From Scratch)





```python
import numpy as np
import matplotlib.pyplot as plt

# Generate larger synthetic dataset with noise
np.random.seed(42)
X = np.linspace(0, 10, 200)  # 200 data points (0-10 years of experience)
true_m = 5.0  # True slope
true_c = 25.0  # True intercept
Y = true_m * X + true_c + np.random.normal(0, 10, len(X))  # Add Gaussian noise

# Initialize params
m = 0.0  # Initial slope guess
c = 0.0  # Initial intercept guess
lr = 0.01  # Learning rate
epochs = 1000  # Training iterations
n = len(X)

costs = []

# Gradient Descent Loop
for i in range(epochs):
    Y_pred = m * X + c
    cost = (1/n) * sum((Y - Y_pred) ** 2)
    costs.append(cost)
    
    # Compute gradients
    dm = (-2/n) * sum(X * (Y - Y_pred))
    dc = (-2/n) * sum(Y - Y_pred)
    
    # Update parameters
    m -= lr * dm
    c -= lr * dc

print(f"Final m (slope): {m:.2f} (True slope = {true_m})")
print(f"Final c (intercept): {c:.2f} (True intercept = {true_c})")

# Plot results
plt.figure(figsize=(12, 5))

# Plot 1: Regression Line
plt.subplot(1, 2, 1)
plt.scatter(X, Y, alpha=0.5, label='Actual Data')
plt.plot(X, m*X + c, color='red', linewidth=2, label='Predicted Line')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)

# Plot 2: Cost vs. Epochs
plt.subplot(1, 2, 2)
plt.plot(range(epochs), costs, color='green')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Cost Reduction During Training')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### ‚úÖ Final Result

The algorithm has learned the best line that minimizes the error between predicted and actual salary based on experience.
