# Practice 02 – Operators-Variable

#### Task 01
Input three values, calculate and print their average:  

**Sample Run:**
``` powershell
Enter first number: 4
Enter second number: 5
Enter third number: 9
Average: 6
```

#### Task 02
Input two numbers. Print their sum, difference and product.  

**Sample Run:**
``` powershell
Enter first number: 6
Enter second number: 4
Sum: 10
Difference: 2
Product: 24
```

**Sample Run:**
``` powershell
Enter Hours: 4

```

#### Task 04
Input previous month budget of an organization for entertainment, which includes cost of milk, tea,
sugar and biscuits. Given there is 20% increase in prices. Calculate and print current month budget.

**Sample Run:**
``` powershell
Previous month budget
Milk:       2000
Sugar:      500
Tea:        1000
Biscuits:   2000

Next month budget
Milk:       2400
Sugar:      600
Tea:        1200
Biscuits:   2400
-----------------------------------------
Total:      6600
```
