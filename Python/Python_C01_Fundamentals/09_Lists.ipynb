# Lists
Lists in Python are collections of indexed items, mutable and capable of holding elements of different data types. In other programming languages, arrays are used instead of lists but both are mostly similar.

**Syntax:**
```python
list_name = [item1, item2, item3, ...]
print(type(list_name))    # <class 'list'>
```

````python
fruits = ['mango', 'apple', 'banana', 'cherry', 'date', 'orange']
print(fruits)
````

## 1. Accessing elements from lists

````python
print(fruits[0])  # First element
print(fruits[1])  # Second element
print(fruits[5])  # Last element
````

1.1 Accessing from reverse order

````python
print(fruits[-1])  # Last element
print(fruits[-2])  # Second to last element
print(fruits[-3])  # Third to last element
print(fruits[-6])  # First element
````

## 2. Mutable OR Changeable
List is Mutable. It means we can change elements of a list in these 3 ways:
- Adding elements
- Removing elements
- Changing elements

For example changing a fruit to "grape"

````python
fruits[1] = 'grape'  # Changing the second element to 'grape'
print(fruits)  # Display the updated list
````

#### 3. Useful functions  
   i. len(my_list)  
   ii. sum(my_list)  
   iii. my_list.append(val)  
   iv. my_list.sort()  
   v. my_list.pop(index)  
   vi. my_list.insert(index, element)  
   vii. my_list.index(element)  
For further reference, you can visit this [link](https://docs.python.org/3.11/tutorial/datastructures.html).

**Note:** 
- `len()` and `sum()` are general functions and not just restricted to lists; they can be used with other data structures like tuples and sets as well. Thats why, they are written like `print()` and `input()` functions.
- Other methods mentioned in the list are specific to lists and thats why they are written like `my_list.append()` or `my_list.sort()`, indicating that they are methods (or functions) of the lists.

````python
# i. len(my_list)
print(len(fruits))  # Length of the list
````

`sum()` method only works with lists containing numeric values. If you try to use it on a list with non-numeric values, it will give error but definitely try it and see the error message.

````python
# ii. sum(my_list)
numbers = [34,23,56,89,45,67]
sum_of_numbers = sum(numbers)
print(sum_of_numbers)  # Sum of the numbers in the list
````

````python
# iii. my_list.append(val)    # Adds another element at the end of the list
fruits.append('kiwi')  # Adding 'kiwi' to the end of the list
print(fruits)  # Display the updated list
````

````python
# iv. my_list.sort()
fruits.sort()  # Sorting the list in alphabetical order
print(fruits)  # Display the sorted list

# > Try it over numbers list
````

````python
# v. my_list.pop(index)
a = fruits.pop(2)  # Removing the third element (index 2)
print(fruits)  # Display the list after removal
print("Removed element:", a)  # Display the removed element
````

````python
# vi. my_list.insert(index, element)
fruits.insert(2, 'pear')  # Inserting 'pear' at index 2
print(fruits)  # Display the list after insertion
````

````python
# vii. my_list.index(element)      
index_of_banana = fruits.index('banana')  # Finding the index of 'banana'
print("Index of 'banana':", index_of_banana)  # Display the index of
````

#### 4. Iterating over lists

````python
# Using while loop
i = 0
while i < len(fruits):
    print(fruits[i])
    i = i + 1
````

> We use this range function way, when we want to have a full control and wants to do some operations on the index of the list. Otherwise, for normal iteration, we can use the `for` loop directly on the list.


````python
# For loop using range function
for i in range(len(fruits)):
    print(fruits[i])
````

````python
# For loop using simple iteration
for i in fruits:
    print(i)
````

#### 5. List slicing
**Syntax:** 
```python 
sliced_list = my_list[start_index:end_index:step]  
```  

If we dont give start_index and end_index, their default the values are as follows:
- start_index = 0
- end_index = len(my_list)   # up to end 
- step = 1

````python
my_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
# sliced_list = my_list[start_index:end_index:step]
new_list = my_list[1:2] 
print(new_list)
````

````python
print(my_list[6:8:3])
````

````python
print(my_list[::-1])    # can be used to reverse the list
````

In case of -ve steps, it will work in opposite way

````python
print(my_list[8:3:-2])   
````
